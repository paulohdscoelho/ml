### Tratamento dos dados

Para que os dados de tuítes fossem tratados de maneira adequada pelos classificadores de ML foi necessário realizar um tratamento sobre os dados a fim de se criar dados normalizados e em um formato que permitisse o treino e teste. Abaixo segue uma descrição sumarizada do que foi feito:

**WordNetLemmatizer**: Lemmatização é o processo de agrupar as diferentes formas flexionadas de uma palavra para que possam ser analisadas como um único item. A lematização é semelhante à derivação, mas traz contexto para as palavras, para isso, liga palavras com significados semelhantes a uma palavra.
Para o presente trabalho foi utilizada a função WordNetLemmatizer da biblioteca NLTK, que realizou a lematização para cada palavra de todos os tuítes, gerando então uma base de dados de palavras lematizadas, ou seja, prontas para análise.

**TfidfVectorizer** Na recuperação de informações, tf–idf (também TF*IDF, TFIDF, TF–IDF ou Tf–idf), abreviação de term frequency–inverse document frequency, é uma estatística numérica que se destina a refletir a importância de uma palavra para um documento em uma coleção ou corpus.

É frequentemente usado como fator de ponderação em buscas de recuperação de informações, mineração de texto e modelagem de usuários. O valor tf–idf aumenta proporcionalmente ao número de vezes que uma palavra aparece no documento e é compensado pelo número de documentos no corpus que contêm a palavra, o que ajuda a ajustar o fato de que algumas palavras aparecem com mais frequência em geral. tf–idf é um dos esquemas de ponderação de termos mais populares atualmente.

No presente trabalho é realizado o cálculo de TF-IDF para todas as palavras dos tuítes previamente lemmatizados. Esse passo é realizado neste notebook utilizando-se da função TfidfVectorizer presente na biblioteca sklearn.feature_extraction.text. Após os tuítes serem fitados no modelo TF-IDF o texto é convertido para uma matriz de documento-termo contendo os valores de cada termo.

# Resultados

Abaixo são testados sobre os dados diversos modelos de Aprendizado de Máquina, passando do mais simples como o Naive Bayes para servir de baseline até algoritmo em redes neurais específicos para NLP. 

É apresentada uma matriz de confusão para cada modelo. Uma matriz de confusão é 
um layout de tabela específico que permite a visualização do desempenho de um algoritmo, normalmente um aprendizado supervisionado.
Cada linha da matriz representa as instâncias em uma classe real enquanto cada coluna representa as instâncias em uma classe prevista, ou vice-versa.
Abaixo uma representação de Matriz de Confusão:

image.png


O nome deriva do fato de que torna fácil ver se o sistema está confundindo duas classes (ou seja, geralmente rotulando erroneamente uma como outra).

Além da Matriz de Confusão é apresentado um report (chamado Classification Report) contendo os valores de acurácia, precisão, recall, f1-score, média macro e média ponderada. 

A precisão é a razão entre tp / (tp + fp) onde tp é o número de true positivos e fp o número de false positivos. A precisão é intuitivamente a habilidade do classificador de não classificar uma amostra negativa como positiva.

O recall é a razão tp / (tp + fn) onde tp é o número de true positivos e fn o número de falso negativos. Recall é intuitivamente explicada como a habilidade do classificador de encontrar todas as amostras positivas.

O score F-bet pode ser interpretado como a média harmônica ponderada da precisão e do recall, onde F-bet score atinge o melhor valor em 1 e o pior score em 0. O F-bet score pesa recall mais do que precisão por um fator determinado beta. Beta igual à 1 significa que recall e precisão são igualmente importantes. O suporte é o número de ocorrências de cada classe em teste_y.

Ao final do trabalho é apresentada a curva ROC para todos os modelos utilizados.

#Importando os pacotes

```python
#importando as bibliotecas necessárias

import pandas as pd
import numpy as np
from numpy import mean
import seaborn as sns
import matplotlib.pyplot as plt
from scipy import stats
from sklearn import tree, neighbors, naive_bayes, svm, datasets
from sklearn.ensemble import GradientBoostingClassifier, RandomForestClassifier, AdaBoostClassifier
from sklearn.neighbors import KNeighborsClassifier, kneighbors_graph
from sklearn.model_selection import GridSearchCV, StratifiedKFold, ParameterGrid, train_test_split, cross_val_score, LeaveOneOut,KFold
from sklearn.preprocessing import MinMaxScaler, StandardScaler, PolynomialFeatures
from sklearn.metrics import classification_report, confusion_matrix, accuracy_score, roc_curve, roc_auc_score, auc, precision_score, recall_score, f1_score
from sklearn.feature_extraction.text import TfidfVectorizer
from sklearn.neural_network import MLPClassifier
from sklearn.linear_model import LogisticRegression
from sklearn.pipeline import Pipeline
from sklearn.cluster import KMeans
from sklearn.decomposition import PCA
from sklearn.manifold import TSNE

import warnings
warnings.filterwarnings('ignore')

%matplotlib inline
```

Lendo os dados e realizando divisão entre dados de treino e de teste utilizando o train_test_split com o parâmetro test_size = 0.3, o que representa que 30% dos dados serão separados para teste e 70% para treino.

Nessa secão também é realizada a transformação dos dados de treino e teste no eixo x para um valor de TfIdf, conforme foi apresentado acima. Serão os dados vetorizados em TfIdf que irão ser alimentados aos classificadores.

```python
#Carregando e tratando os dados

SEED = 500

np.random.seed(SEED)
url = "https://raw.githubusercontent.com/paulohdscoelho/ml/master/TP_final/sentimentos_lemmatizados.csv"
corpus = pd.read_csv(url,sep="\t",encoding='latin-1')

X, Y = corpus['text_final'], corpus['sentimento']

treino_x, teste_x, treino_y, teste_y = train_test_split(corpus['text_final'],corpus['sentimento'],test_size=0.3)

Tfidf_vect = TfidfVectorizer(max_features=5000)
Tfidf_vect.fit(corpus['text_final'])
treino_x_Tfidf = Tfidf_vect.transform(treino_x)
teste_x_Tfidf = Tfidf_vect.transform(teste_x)

target_names = ['class 0', 'class 1']
```

# Plot de valores de TFIDF de acordo com os termos

Nesta seção realizei a plotagem dos dados de treino e teste x com TFIDF em um gráfico de scatter plot. Basicamente os resultados mostram que há uma correlação grande entre o valor do termo e do TfIdf correspondente ao termo. Notadamente há uma concentração dos pontos nos pontos próximos de (0,0).


```python
num_clusters = 10
num_seeds = 10
max_iterations = 300

# create k-means model with custom config
clustering_model = KMeans(
    n_clusters=num_clusters,
    max_iter=max_iterations        
)

X = treino_x_Tfidf.todense()

pca = PCA(n_components=2).fit(X)
data2D = pca.transform(X)

plt.rcParams["figure.figsize"] = (18,12)
scatter = plt.scatter(data2D[:,0], data2D[:,1], c=treino_y)
plt.title("Termo vs valor de TFIDF")
plt.xlabel('Termo')
plt.ylabel('TFIDF')
plt.legend(handles=scatter.legend_elements()[0], labels=['0','1'])
plt.show() 
```

# Classificador Naive Bayes

O Naive Bayes será usado como o baseline do experimento, por isso não vou variar o hiperparâmetro do algoritmo que é o var_smoothing.
No report de classificação fica evidente que o modelo possui uma acurácia ok de 0.76 e um recall e f1-score altos para a classe 0 e baixos para a classe 1.

A matriz de confusão apresenta valores altos para TP e baixos para os erros, o que indica que o modelo prevê com mais corretude os dados.

```python
# Classifier - Algorithm - Naive Bayes
Naive = naive_bayes.MultinomialNB()
Naive.fit(treino_x_Tfidf,treino_y)# predict the labels on validation dataset
predictions_NB = Naive.predict(teste_x_Tfidf)# Use accuracy_score function to get the accuracy

print(classification_report(teste_y, predictions_NB, target_names=target_names))

cm = confusion_matrix(teste_y, predictions_NB)
plt.figure(figsize = (10,7))
print("Confusion Matrix:")
sns.heatmap(cm, annot=True,fmt='g')
```

# Classificador Regressão Logística

A Regressão Logística será o próximo algoritmo a ser treinado. Como pode ser verificado abaixo, a acurácia do modelo foi de 78%. Além disso, os valores da matriz de confusão apresentam valores altos para o TP e o TN, o que indica que o modelo performa bem.

```python
# Classifier - Algorithm - Logistic Regression
clf_reglog = LogisticRegression(max_iter = 100).fit(X = treino_x_Tfidf, y = treino_y)
predictions_reglog = clf_reglog.predict(teste_x_Tfidf) # Use accuracy_score function to get the accuracy

print(classification_report(teste_y, predictions_reglog, target_names=target_names))

cm = confusion_matrix(teste_y, predictions_reglog)
plt.figure(figsize = (10,7))
print("Confusion Matrix:")
sns.heatmap(cm, annot=True,fmt='g')
```

# Classificador Decision Tree

O próximo algoritmo a ser treinado é a Árvore de Decisão. Os hiperparâmetros que serão testados é a profundidade máxima da árvore e o peso de cada uma das classes.

No fim estão os resultados das Árvores de Decisão. Veja que os hiperparâmetros que tiveram melhor acurácia média em teste foi com profundidade máxima igual a None e sem peso em nenhuma das classes.
A Acurácia Média ficou em torno de 0.68, o mesmo valor apresentado pelo classification report abaixo. Tanto o precision, o recall e o f1-score são mais altos para a classe 0 no mesmo report.

A seguir é apresentada a Matiz de Confusão para os dados. Nota-se que os valores de True Positive e True Negative são mais altos do que os demais, o que significa que o modelo prevê corretamente os dados.

O Hiperparâmetro que obteve melhor valor de mean test accuracy foi para o valor de max_depht None e class weight None na árvore, ou seja, com altura infinita e sem pesos. 

```python
#Aproximadamente 4 minutos para rodar

pipeline_arvore_decisao = Pipeline(steps= [('arvore_decisao', tree.DecisionTreeClassifier(criterion = 'entropy'))])

# definindo os valores possiveis para o hiperparametro profundidade da árvore
hiperparametro_arvore_decisao = {'arvore_decisao__max_depth': [1, 2, 10, 40, 50, 80, 100, None],
                                 'arvore_decisao__class_weight': [None, 'balanced', {'H' : 1, 'D': 10, 'A': 10}]}

clf_arvore_decisao = GridSearchCV(estimator = pipeline_arvore_decisao, param_grid = hiperparametro_arvore_decisao, 
                                  cv = StratifiedKFold(n_splits = 5, shuffle = True, random_state = SEED),
                                  n_jobs = -1, scoring = ['accuracy', 'f1_macro'], refit = 'f1_macro', 
                                  return_train_score = True, verbose = True)

# ajustando o algoritmo aos dados do nosso estudo
clf_arvore_decisao.fit(X = treino_x_Tfidf,
                        y = treino_y)
```

```python
#Resultados Decision Tree
pd.DataFrame(clf_arvore_decisao.cv_results_).drop(['params', 'split0_test_accuracy',	'split1_test_accuracy',	'split2_test_accuracy',	'split3_test_accuracy',	'split4_test_accuracy',
                                                   'split0_train_accuracy',	'split1_train_accuracy',	'split2_train_accuracy',	'split3_train_accuracy',	'split4_train_accuracy',
                                                    'split0_test_f1_macro',	'split1_test_f1_macro',	'split2_test_f1_macro',	'split3_test_f1_macro',	'split4_test_f1_macro',
                                                    'split0_train_f1_macro',	'split1_train_f1_macro',	'split2_train_f1_macro',	'split3_train_f1_macro',	'split4_train_f1_macro'], axis = 1).sort_values('rank_test_f1_macro', ascending = True)
```

```python
#Report de classificação e matriz de confusão
predictions_dt = clf_arvore_decisao.predict(teste_x_Tfidf)
acuracia = accuracy_score(teste_y, predictions_dt) * 100

print(classification_report(teste_y, predictions_dt, target_names=target_names))

cm = confusion_matrix(teste_y, predictions_dt)
plt.figure(figsize = (10,7))
print("Confusion Matrix:")
sns.heatmap(cm, annot=True,fmt='g')
```

Os hiperparâmetros que obtiveram melhor desempenho foram class weight e max depht None e class weight igual a	200, ou seja, o classificador sem peso na classe, com profundidade infinita na árvore e com o maior número de estimators obteve o melhor desempenho.

A seguir é apresentada a matriz de confusão e observa-se que o classificador retorna valores altos para TP e TN, maior do que os valores de erro FP e FN que alcançam valores próximos. 

```python
#aproximadamente 12 minutos para rodar
#classificador Random Forest

pipeline_floresta_aleatoria = RandomForestClassifier(criterion = 'entropy')

# definindo os valores possíveis para os hiperparâmetros.
hiperparametro_floresta_aleatoria = {'n_estimators': [1, 10, 20, 50, 100, 200], 
                                     'max_depth': [1, 2, 10, 20, 50, None],
                                    'class_weight': [None, 'balanced', {'H' : 1, 'D': 10, 'A': 10}]}

# criando o grid que irá testar todos esses números de árvores da floresta e retornar
# o melhor a partir da avaliação dos resultados através de validação cruzada.
clf_floresta_aleatoria = GridSearchCV(estimator = pipeline_floresta_aleatoria, 
                                       param_grid = hiperparametro_floresta_aleatoria, 
                                       cv = StratifiedKFold(n_splits = 5, shuffle = True, random_state = SEED), 
                                       n_jobs = -1, scoring = ['accuracy', 'f1_macro'], refit = 'f1_macro', 
                                       return_train_score = True, verbose = True)

# ajustando o algoritmo da floresta aleatória aos dados do nosso estudo
clf_floresta_aleatoria.fit(X = treino_x_Tfidf, y = treino_y)
```

```python
#Exibindo dados de random forest
pd.DataFrame(clf_floresta_aleatoria.cv_results_).drop('params', axis = 1).drop(['split0_test_accuracy',	'split1_test_accuracy',	'split2_test_accuracy',	'split3_test_accuracy',	'split4_test_accuracy',
                                                                                'split0_train_accuracy',	'split1_train_accuracy',	'split2_train_accuracy',	'split3_train_accuracy',	'split4_train_accuracy',
                                                                                'split0_test_f1_macro',	'split1_test_f1_macro',	'split2_test_f1_macro',	'split3_test_f1_macro',	'split4_test_f1_macro',
                                                                                'split0_train_f1_macro',	'split1_train_f1_macro',	'split2_train_f1_macro',	'split3_train_f1_macro',	'split4_train_f1_macro'], axis = 1).sort_values('rank_test_f1_macro', ascending = True)
```

```python
#Exibindo report de classificação e matriz de confusão
predictions_rf = clf_floresta_aleatoria.predict(teste_x_Tfidf)
acuracia = accuracy_score(teste_y, predictions_rf) * 100

print(classification_report(teste_y, predictions_rf, target_names=target_names))

cm = confusion_matrix(teste_y, predictions_rf)
plt.figure(figsize = (10,7))
print("Confusion Matrix:")
sns.heatmap(cm, annot=True,fmt='g')
```

# Classificador MLP (Multi-layer Perceptron)

Multilayer Perceptron (MLP) é uma classe full conectada de rede neural artificial de antecipação (ANN). O termo MLP é usado de forma ambígua, às vezes significando qualquer ANN.
Uma MLP consiste de pelo menos três camadas de nós de entrada, cada nó é um neurônio que usa backpropagation para treinar. 
As suas múltiplias camadas e a função de ativação não linear diferencia MLP de um perceptron linear.

Para o presente trabalho o MLP retornou uma acurácia de 75% com valores altos de TP e TN na matriz de confusão, o que indica que o modelo prevê de forma satisfatória os valores.

```python
# Classifier - Algorithm - MLP
mlp_clf = MLPClassifier(solver='lbfgs', alpha=1e-5, hidden_layer_sizes=(5, 2), random_state=SEED, max_iter=1000).fit(treino_x_Tfidf, treino_y)
predictions_MLP = mlp_clf.predict(teste_x_Tfidf)# Use accuracy_score function to get the accuracy

print(classification_report(teste_y, predictions_MLP, target_names=target_names))

cm = confusion_matrix(teste_y, predictions_MLP)
plt.figure(figsize = (10,7))
print("Confusion Matrix:")
sns.heatmap(cm, annot=True,fmt='g')
```

# Curva ROC

A Curva Característica de Operação do Receptor, ou, do inglês, Receiver Operating Characteristic Curve (ROC curve), ou, simplesmente, curva ROC, é uma representação gráfica que ilustra o desempenho (ou performance) de um sistema classificador binário à medida que o seu limiar de discriminação varia.

Para este trabalho é plotada a curva ROC para cada um dos 7 modelos de aprendizagem de máquina testados.

É evidente de acordo com o plot que o valor de AUC-ROC é maior para os modelos Nayve Bayes, Logistic Regression, e SVM. Dos modelos, Ada Boosting e Decision Tree obtiveram os valores mais baixos dentre todos.

Nota-se que todas as curvas dos sete modelos aqui experimentados tendem a aumentar a taxa de True Positive conforme o valor de False Positive cresce.
